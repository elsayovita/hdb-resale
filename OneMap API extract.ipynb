
# Import Packages

```python
import pandas as pd 
import numpy as np

# ONE MAP API data extraction
import json
import requests
```

# Import dataset

```python
# housing info
df_houseinfo = pd.read_csv('hdb-property-information.csv')
```

# Concatenate block and street name to get ADDRESS
- Address is later used to obtain coordinates for geographical map visualisation

```python
df_houseinfo.insert(0, "address", df_houseinfo["blk_no"] + " " + df_houseinfo["street"])
```

# Generate longitude and latitude from OneMap API:

```python
hdb_lat = []
hdb_long = []
for i in range(len(df_houseinfo["address"])):
    try:
        query_address = df_houseinfo.loc[i, "address"]
        query_string = 'https://developers.onemap.sg/commonapi/search?searchVal='+str(query_address)+'&returnGeom=Y&getAddrDetails=Y'
        resp = requests.get(query_string)
        query_data =json.loads(resp.content)
        hdb_lat.append(query_data["results"][0]["LATITUDE"])
        hdb_long.append(query_data["results"][0]["LONGITUDE"])
        print('Extracting',i)
    except:
        hdb_lat.append("error")
        hdb_long.append("error")
        print('Failed to extract',i)
```

```python
df_houseinfo.insert(3, "latitude", hdb_lat)
df_houseinfo.insert(4, "longitude", hdb_long)
```

# Export table to excel

```python
df_houseinfo.to_excel(r"housing_list.xlsx", index = False)
```
